```python
from langchain.document_loaders import PyPDFLoader
```

```python
# Load PDF
loaders = [
    # Duplicate documents on purpose - messy data
    PyPDFLoader("/content/data/tmp_doc.pdf")
    # PyPDFLoader("/content/documents/ddpm.pdf"),
]
docs = []
for loader in loaders:
    docs.extend(loader.load())
```

```python
# Split
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 200,
    chunk_overlap = 15
)
```

```python
splits = text_splitter.split_documents(docs)
len(splits)
```

```python
splits[0]
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
from langchain.vectorstores import Chroma
```

```python
persist_directory = '/content/docs/chroma/'
```

```python
!rm -rf /content/docs/chroma  # remove old database files if any
```

```python
vectordb = Chroma.from_documents(
    documents=splits,
    embedding=embeddings,
    persist_directory=persist_directory
)
```

```python
print(vectordb._collection.count())
```

```python
question = "What is title?"
docs = vectordb.similarity_search(question,k=3)
len(docs)
```

```python
llm = HuggingFaceHub(repo_id="tiiuae/falcon-7b-instruct", model_kwargs={"max_length":300, "max_new_tokens":300})
```

```python
template = """
You are a helpful bot that makes funny jokes about {topic}

Answer: Let's think step by step.
"""

prompt_template = PromptTemplate.from_template(template=template)
chain = LLMChain(llm=llm, prompt=prompt_template)
chain.run("Ducks")
```

```python
# Build prompt
template = """Use the following pieces of context to answer the question at the end. If you don't know the answer, just say that you don't know, don't try to make up an answer. Use three sentences maximum. Keep the answer as concise as possible. Always say "thanks for asking!" at the end of the answer.
{context}
Question: {question}
Helpful Answer:"""
QA_CHAIN_PROMPT = PromptTemplate.from_template(template)
```

```python
from langchain.chains import RetrievalQA
```

```python
# # Run chain
# qa_chain = RetrievalQA.from_chain_type(
#     llm,
#     retriever=vectordb.as_retriever(),
#     return_source_documents=True,
#     chain_type_kwargs={"prompt": QA_CHAIN_PROMPT}
# )
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm,
    retriever=vectordb.as_retriever(),
    # chain_type="refine",
    chain_type="map_reduce"
)
```

```python
summary_prompt = "what is title of the content"

result = qa_chain({"query": summary_prompt})
result["result"]
```

```python
notes_prompt = "understand all content, and explain it"

result = qa_chain({"query": notes_prompt})
result["result"]
```
